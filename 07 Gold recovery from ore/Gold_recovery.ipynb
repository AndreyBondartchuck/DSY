### Проверим, что эффективность обогащения рассчитана правильно. Вычислим её на обучающей выборке для признака rougher.output.recovery. Найдём MAE между нашими расчётами и значением признака 'rougher.output.recovery'. 

Эффективность обогащения рассчитывается по формуле

**Recovery = C*(F-T)/(F*(C-T)) * 100%**

- C — доля золота в концентрате после флотации/очистки;    -> rougher.output.concentrate_au
- F — доля золота в сырье/концентрате до флотации/очистки; ->  rougher.input.feed_au
- T — доля золота в отвальных хвостах после флотации/очистки. -> rougher.output.tail_au

```python
def calculate_rougher_recovery(raw):
    
    C = raw['rougher.output.concentrate_au']
    F = raw['rougher.input.feed_au']
    T = raw['rougher.output.tail_au']
    
    denominator = F*(C-T)
    if denominator != 0:
        rougher_recovery = (C*(F-T)/denominator) * 100
    else:
        rougher_recovery = 0
        
    return rougher_recovery
```

```python
df_train_copy = df_train.copy()
df_train_copy.dropna(subset = ['rougher.output.concentrate_au',\
                               'rougher.input.feed_au','rougher.output.tail_au','rougher.output.recovery'], inplace = True)

df_train_copy['calculate.rougher.output.recovery'] = df_train_copy.apply(calculate_rougher_recovery, axis=1)     
```

```python
MAE = mean_absolute_error(df_train_copy['rougher.output.recovery'],df_train_copy['calculate.rougher.output.recovery'])
print('MAE между расчётным recovery и представленным значением признака rougher.output.recovery = {:.20f}.'.format(MAE))
```

**Вывод: МАЕ между расчетным показателем recovery и представленным в выборке на этапе rougher, крайне мал и равен  0.00000000000000930342. Что говорит о доверии к данным Recovery представленным на этапе rougher.**

### Посмотрим, как меняется концентрация металлов (Au, Ag, Pb) на различных этапах очистки. Опишем выводы.

```python
# соберем удобный для визуализации датасет
def concentrate(df):
    df_Ag = pd.DataFrame({'rougher':df['rougher.input.feed_ag'],'Metal':'Ag','step':'rougher.input'})
    df_Pb = pd.DataFrame({'rougher':df['rougher.input.feed_pb'],'Metal':'Pb','step':'rougher.input'})
    df_Au = pd.DataFrame({'rougher':df['rougher.input.feed_au'],'Metal':'Au','step':'rougher.input'})
    df_concentrate = pd.concat([df_Ag,df_Pb,df_Au])
    
    df_Ag = pd.DataFrame({'rougher':df['rougher.output.concentrate_ag'],'Metal':'Ag','step':'rougher.output'})
    df_Pb = pd.DataFrame({'rougher':df['rougher.output.concentrate_pb'],'Metal':'Pb','step':'rougher.output'})
    df_Au = pd.DataFrame({'rougher':df['rougher.output.concentrate_au'],'Metal':'Au','step':'rougher.output'})
    df_concentrate = pd.concat([df_concentrate,df_Ag,df_Pb,df_Au])
    
    df_Ag = pd.DataFrame({'rougher':df['primary_cleaner.output.concentrate_ag'],'Metal':'Ag','step':'primary'})
    df_Pb = pd.DataFrame({'rougher':df['primary_cleaner.output.concentrate_pb'],'Metal':'Pb','step':'primary'})
    df_Au = pd.DataFrame({'rougher':df['primary_cleaner.output.concentrate_au'],'Metal':'Au','step':'primary'})
    df_concentrate = pd.concat([df_concentrate,df_Ag,df_Pb,df_Au])
    
    df_Ag = pd.DataFrame({'rougher':df['final.output.concentrate_ag'],'Metal':'Ag','step':'final'})
    df_Pb = pd.DataFrame({'rougher':df['final.output.concentrate_pb'],'Metal':'Pb','step':'final'})
    df_Au = pd.DataFrame({'rougher':df['final.output.concentrate_au'],'Metal':'Au','step':'final'})
    df_concentrate = pd.concat([df_concentrate,df_Ag,df_Pb,df_Au])
    
    return df_concentrate

df_concentrate = concentrate(df_full)
```

```python
plt.figure(figsize=(12, 5))
ax = sns.boxplot(x=df_concentrate['step'], y=df_concentrate['rougher'], hue=df_concentrate['Metal']);
plt.title("Концентрация металлов на различных этапах",fontsize=15, weight = 'bold')
plt.ylabel('Концентрация',fontsize=15, weight = 'bold')
plt.xlabel('Этапы очистки',fontsize=15, weight = 'bold');
```

### Очистка данных от выбросов

```python
# функция очищает датасет от выбросов, отбрасывая объекты за пределами межквартильного размаха
def df_cleaner(df, column_list=[], query=False, koeff=1.5):
    if len(column_list)==0:
        column_list = list(df.columns)
    for column in column_list:
        q25 = df[column].quantile(0.25)                 
        q75 = df[column].quantile(0.75) 
        minimum = df[column].min()
        maximum = df[column].max()    
        irq = np.abs(q75 - q25)    
        left = q25 - koeff*irq
        right = q75 + koeff*irq
        if query: 
            sql_sentence='@left <= '+ column + ' and '+ column + ' <= @right'
            df = df.query(sql_sentence)
        else:
            df_tmp = df[(df[column]>left)&(df[column]<right)]
            if len(df_tmp) != 0:
                df=df_tmp.copy()
        pass
    return df
```

#### Метрика sMAPE

```python
# функция расчета метрики
def smape(y_true, y_pred):
    result=0
    if len(y_true) != len(y_pred):
        return -1
    for i in range(len(y_true)):        
        denominator = (np.abs(y_true[i]) + np.abs(y_pred[i]))/2
        if denominator == 0:
            continue
        result += np.abs(y_true[i] - y_pred[i]) / denominator
    result = result/len(y_true)*100
    return result
```

Посмотрим, предварительно, значениz метрик на 'рандомной модели'. В качестве предсказания возьмем средние по столбцу **rougher.output.recovery** и **final.output.recovery**.

```python
rougher_recovery_mean = rougher_train['rougher.output.recovery'].mean()

y_true = rougher_train['rougher.output.recovery']
y_pred = [rougher_recovery_mean]*len(y_true)

smape_rougher = smape(y_true, y_pred)
print('Значение метрики sMAPE для rougher на рандомной модели {}'.format(smape_rougher))
```

```python
final_recovery_mean = final_train['final.output.recovery'].mean()

y_true = final_train['final.output.recovery']
y_pred = [final_recovery_mean]*len(y_true)

smape_final = smape(y_true, y_pred)
print('Значение метрики sMAPE для final на рандомной модели {}'.format(smape_final))
```

```python
total_smape = 0.25 * smape_rougher + 0.75 * smape_final
print('Значение итоговой метрики sMAPE на рандомной модели {}'.format(total_smape))
```

#### Модель простой линейной регрессии

Напишем функцию для подстановки различных вариантов моделей для линейной регрессии

```python
result_list = []

def research_model(model, model_name =''):
    
    print(model_name)
    
    ################################################################################
    # rougher
    features = rougher_train.drop(['rougher.output.recovery'], axis=1)
    target = rougher_train['rougher.output.recovery'] 
    
    # model = LinearRegression() 

    scoring = {'smape': make_scorer(smape)}
    scores = cross_validate(model, features, target, cv=5, scoring=scoring, return_train_score=True, return_estimator=True)
    print()
    print('############################## rougher ##################################')
    print('cross_validate')
    print('Scores smape = '+ str(scores['test_smape']))
    
    model.fit(features,target)
    
    # получим предсказания модели на отложенной тестовой выборке
    features_test = rougher_test.drop(['rougher.output.recovery'], axis=1)
    target_test = rougher_test['rougher.output.recovery'] 

    predictions_test = model.predict(features_test) 

    # посчитаем значение метрики smape на тестовой выборке
    rougher_smape = smape(target_test, predictions_test)
    print()
    print("Метрика sMAPE(rougher) модели на тестовой выборке:", rougher_smape)

    ################################################################################
    # final

    features = final_train.drop(['final.output.recovery'], axis=1)
    target = final_train['final.output.recovery'] 

    # model = LinearRegression() 

    scoring = {'smape': make_scorer(smape)}
    scores = cross_validate(model, features, target, cv=5, scoring=scoring, return_train_score=True, return_estimator=True)
    print()
    print('############################## final ##################################')
    print('cross_validate')
    print('Scores smape = '+ str(scores['test_smape']))
    
    # получим предсказания наилучшей модели на отложенной тестовой выборке
    features_test = final_test.drop(['final.output.recovery'], axis=1)
    target_test = final_test['final.output.recovery'] 

    model.fit(features,target)
    
    predictions_test = model.predict(features_test) 

    # посчитаем значение метрики smape на тестовой выборке
    final_smape = smape(target_test, predictions_test)
    print()
    print("Метрика sMAPE(final) модели на тестовой выборке:", final_smape)

    # итоговая метрика
    total_smape = 0.25 * rougher_smape + 0.75 * final_smape
    
    print()
    print('Итоговая метрика sMAPE = {}'.format(total_smape))
    
    result_list.append({ 'Model':model_name,'sMAPE_rougher':rougher_smape, \
                        'sMAPE_final':final_smape, 'sMAPE_total':total_smape})
    
```
